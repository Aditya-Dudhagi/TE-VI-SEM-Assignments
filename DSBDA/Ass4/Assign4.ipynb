```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_names = sns.get_dataset_names()
for i in dataset_names:
    print(i)
```

```python
df = sns.load_dataset('titanic')
df
```

### CountPlot 

```python
sns.countplot(data=df, x = "survived")
plt.title("Survival Count")
plt.show()

# more people did not survive than survived
```

### Bar plot : survival by gender

```python
sns.barplot(data=df, x="sex", y="survived")
plt.title("Survival Rate by Gender")
plt.show()

# female had higher survival rate
```

```python
# Box plot to show age distribution based on survival
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="survived", y='age')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Younger passengers had better survival chances.
```

```python
# Box plot Age by Sex
# Plot box plot
sns.boxplot(data=df, x="sex", y="age", hue="survived")

# Labels and title
plt.title("Distribution of Age by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()


# Observations:
# The median age of survivors is lower compared to non-survivors for both genders.
# Female passengers show a higher survival rate across most age groups.
# Among males, younger passengers had a better chance of survival than older ones.
# The presence of outliers indicates that a few elderly passengers survived, but they are rare.
# Overall, the plot suggests that gender and age significantly influenced survival, with women and children having higher survival chances.
```

```python
# Bar plot to show how Class affected Survival
sns.barplot(data=df, x="pclass", y="survived")
plt.title("Survival Rate by Passenger Class")
plt.show()

#1st class passengers survived more.
```

```python
# Scatter plot showing relationship between age and fare
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='age', y='fare', hue='survived')
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# line plot Survival rate by Passenger Class
sns.lineplot(data=df, x="pclass", y="survived", markers="o")
plt.title("Survival rate by passenger class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# Line Plot: Shows trend of survival rate across classes
# Survival rate decreases as passenger class goes from 1st → 3rd.
```

```python
# Histogram showing age distribution
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='age', bins=30)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# violin plot 
sns.violinplot(data=df, x="survived", y="age")
plt.title("Age distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.show()
# Younger passengers had a higher chance of survival.
```

```python
# strip plot: Fare vs Survival
sns.stripplot(data=df, x="survived", y="fare", jitter=True)
plt.title("Fare vs Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.show()

# Passengers who paid higher fares were more likely to survive.
```

### Histogram showing ticket fare distribution


```python
# Plot histogram of fare
sns.histplot(data=df, x="fare", bins=30, kde=True)

# Labels and title
plt.title("Distribution of Ticket Prices (Fare)")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```
