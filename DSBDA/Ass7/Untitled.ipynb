```python
### import Libraries

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

### sample data

```python
text = """Natural Language Processing is a fascinating field of Artificial Intelligence.
It helps computers understand human language."""
```

### Tokenization:

#### Tokenization is the process of breaking a text document into smaller units called tokens.
#### Tokens can be words, sentences, or characters.

```python
tokens = word_tokenize(text)
print("Tokens: " , tokens)
```

### Part Of Speech Tagging

#### POS Tagging is the process of assigning grammatical labels to each word in a sentence, such as noun, verb, adjective, etc

```python
pos_tags = pos_tag(tokens)
print("POS Tags: ", pos_tags)
```

### Remove Stop Words:

#### Stopword removal is the process of removing common words that do not add significant meaning to the text, such as is, the, and, in, of, etc

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("After Stopword Removal: ", filtered_tokens)
```

### Stemming:

#### Stemming is the process of reducing words to their root or base form by removing suffixes.

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed words: ", stemmed_words)
```

### Lemmatization:

#### Lemmatization is the process of reducing a word to its base or dictionary form (called lemma) using linguistic rules.

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized words: ", lemmatized_words)
```

## PART 2: TERM FREQUENCY AND INVERSE DOCUMENT FREQUENCY

### Term Frequency:

#### Term Frequency measures how many times a word appears in a document. It shows the importance of a word within that specific document.




#### TF = (Number of times a word appears in a document / Total number of words in the document)


```python
count_vectorizer = CountVectorizer()
X_counts = count_vectorizer.fit_transform([text])


print("Term Frequency: ")
print(count_vectorizer.get_feature_names_out())
print(X_counts.toarray())
```

### IDF:

#### Inverse Document Frequency measures how important a word is across multiple documents. Words that appear in many documents (like "the") get lower IDF scores.

### IDF = LOG ( Total number of documents / Number of documents containing the word )

#### TF-IDF: TF-IDF is a numerical statistic that combines TF and IDF to measure how important a word is in a document relative to a collection of documents.

### FORMULA: TF-IDF = TF * IDF

```python
tfidf_vectorizer =  TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform([text])


print("TF-IDF Feature Names:")
print(tfidf_vectorizer.get_feature_names_out())

print("TF-IDF Values:")
print(X_tfidf.toarray())
```
