```python
bins = [0, 20, 30, 45, 60, 100]
```

```python
labels = ['Teenage', 'Yound', 'Middle_Age', 'Adult', 'Old']
```

```python
df['Age_Range'] = pd.cut(df['Age'], bins = bins, labels = labels, right = False)
```

```python
grouped = df.groupby('Age_Range')['Spending Score (1-100)']
```

```python
max_vals = grouped.max()
```
