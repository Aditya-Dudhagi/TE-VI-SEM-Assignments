### detecting outliers using 3) Z-Score

```python
# z-score is called standard score. 
```

```python
# helps to understand how far is data point from the mean.
#z-score values of data points can be utilized to define the outliers.
```

#### Zscore = (data_point - mean)/ std. deviation

```python
import numpy as np
from scipy import stats
```

```python
# now calculate the absolute Z-score of the math score column to identify outliers.
```

```python
z = np.abs(stats.zscore(df['math score']))
```

```python
print(z)
```

```python
# define an outlier threshold value
```

```python
threshold = 2.5
```

```python
# finding the outliers

# values far from the mean have higher Z-scores and are treated as outliers 
# when Z-score exceeds a threshold
```

```python
sample_outliers = np.where(z>threshold)
```

```python
sample_outliers
```

```python
not_outliers = np.where(z<threshold)
not_outliers
```

### Detecting outliers using 4) Inter Quartile Range (IQR) 

```python
# this is most commonly used and most trusted approach used in research field
```

```python
sorted_rscore = sorted(df['reading score'])
```

```python
# sort reading score feature and store it into sorted_rscore
```

```python
sorted_rscore
```

```python
# calculate and print Quartile 1 and Quartile 3
```

```python
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
```

```python
# np.percentile(..., 25) finds the value below which 25% of the data lies.
# np.percentile(..., 75) finds the value below which 75% of the data lies.
```

```python
print(q1, q3)
```

```python
# calculate value of iqr
iqr = q3 - q1
```

```python
# calc and print upper and lower bound to define the outlier base value
lwr_bound = q1 - (1.5 * iqr)
upr_bound = q3 + (1.5 * iqr)
print(lwr_bound, upr_bound)
```

```python
# print outliers

r_outliers = []
for i in sorted_rscore:
    if (i<lwr_bound or i>upr_bound):
        r_outliers.append(i)
print(r_outliers)
```

### Handling of Outliers

#### 1. Trimming / removing the outlier

#### 2. Quantile based flooring and capping

```python
# in this technique, the outlier is capped at a certain value above the 90th percentile value
#or floored at a factor below the 10th percentile value
```

```python
df = pd.read_csv('outliers.csv')
```

```python
df_stud = df
```

```python
nin_perc =  np.percentile(df_stud['math score'], 90)
b = np.where(df_stud['math score']>nin_perc, nin_perc, df_stud['math score'])
print("New array: ", b)
```

```python
# A new column called m score is inserted as the second column, containing the modified math scores.
```

```python
df_stud.insert(1, "m score", b, True)
df_stud
```

#### 3. mean/median imputation

```python
# replace the outliers with the median values since mean is affected by the outliers
```

```python

df.boxplot('reading score')
```

```python
plt.show()
```

```python
median = np.median(sorted_rscore)
```

```python
median
```

```python
#replace the upper bound outliers using median value
```

```python
refined_df = df
refined_df['reading score'] = np.where(refined_df['reading score'] > upr_bound, median, refined_df['reading score'])
```

```python
# replace lower bound using median value
refined_df['reading score'] = np.where(refined_df['reading score']<lwr_bound, median, refined_df['reading score'])
refined_df
```
