## Convert the variables to proper data types

```python
# convert Survived and Pclass to categorial
```

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
df['Pclass'] = df['Pclass'].astype('category')
```

```python
# convert Sex and Embarked to category
```

```python
df['Sex'] = df['Sex'].astype('category')
```

```python
df['Embarked'] = df['Embarked'].astype('category')
```

```python
df.dtypes
```

### Handle missing values

```python
# fill missing age with mean and embarked with mode
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

### Convert Categorical Variables into Quantitative Variables

```python
# Label Encoding ( Binary Categories ) 
# map male to 0 and female to 1
#convert gender into numerical form which is required for ML algorithms
```

```python
df['Sex'] = df['Sex'].map({'male' : 0, 'female' : 1})
```

```python
df.head(10)
```

```python
df.tail()
```

```python
# One-hot Encoding 
#converts categorical variable into dummy variables
# drop_first=True avoids multicollinearity
```

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```
